```python
import pandas as pd
import random

# read the data from the downloaded CSV file.
data = pd.read_csv('https://s3-eu-west-1.amazonaws.com/shanebucket/downloads/uk-500.csv')
# set a numeric id for use as an index for examples.
data['id'] = [random.randint(0,1000) for x in range(data.shape[0])]

data.head()
```

# .iloc

```python
#Single selections using iloc an DataFrame
data.iloc[0]
# first row of data frame
```

```python
#Select last row using iloc
data.iloc[-1]
```

```python
#Select second row
data.iloc[1]
```

```python
#la primera columna. el primer parámetro indica que quiero todas las filas (:)
#el segundo parámetro indica que quiero la primera columna
#con loc debería poner el nombre de la columna como string, en lugar del 0
#si pongo el 0 entre corchetes el look es como dataframe
data.iloc[:,0]
```

```python
#segunda columna del DF

data.iloc[:,1]
data.loc[:,'last_name']
```

```python
#última columna
data.iloc[:,-1]
```

```python
#las primeras 5 filas del df
data.iloc[:5]
```

```python
#primeras dos columnas del df con todas las filas
data.iloc[:,:2]
```

```python
#1a, 4a, 7a, 25a fila + 1a, 6a, 7a columna
data.iloc[[0,3,6,24], [0,5,6]]
```

```python
#las primeras 5 filas y las columns 5,6 y 7
data.iloc[:5, 4:7]
```

# .loc

```python
# primero vamos a cambiarle el índice

data.set_index('last_name', inplace=True)
data.head()
```

```python
data.head()
```

```python
#queremos de Zigomalas a Veness
data.loc['Zigomalas':'Veness']
```

```python
#queremos la fila del señor andrade
data.loc['Andrade']
```

```python
#filtrar a andrade y vanessa y columnas city y county
data.loc[['Andrade', 'Veness'], ['city', 'county']]
```

```python
#los mismos, con first name, address and city columns
data.loc[['Andrade', 'Veness'], ['first_name', 'address', 'city']]
```

```python
#change the index to be set on the id column
data.set_index('id', inplace=True)
```

```python
data.head()
```

```python
# select the row with 'id' = 952

data.loc[952]
```

```python
#queremos los elementos de first_name que sean Antonio

data.loc[data['first_name']=='Antonio']
```

```python
#sin el loc es lo puto mismo

data[data['first_name']=='Antonio']
```

```python
#filas cuyo nombre sea aantonio y todas las columnas entre city y mail

data.loc[data['first_name']=='Antonio', 'city':'email']
```

```python
#selecciona las filas donde mail column ends with hotmail.com, include all columns
data.loc[data['email'].str.endswith('hotmail.com')]
```
