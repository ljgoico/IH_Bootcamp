```python
#pie chart

labels = ['Vegetarians', 'Non-Vegetarians', 'Vegans']
sizes = [40, 45, 15]
```

```python
#esto es lo que hace que el pie salga separado
explode = (0,0.1,0)

#enga
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels = labels, autopct='%1.1f%%', shadow=True, startangle=90)
#autopct='%1.1f%%' esto es para que muestre los pocentajes
plt.show()
```

## 2. Seaborn

```python
import seaborn as sns
sns.set()

import pandas as pd
```

```python
data = pd.DataFrame()

data['X'] = np.arange(0,10,1)
data['Y'] = data['X']**2
```

```python
data
```

```python
sns.lineplot(x=data.X,y=data.Y)
```

```python
#gráfico que tiene que ver con la regresión

sns.lmplot('X', 'Y', data=data)
```

```python
data = pd.read_csv('/Users/ljgoico/desktop/ironhack/module-2/week-10/vehicles.csv')
```

```python
sns.scatterplot(x='CO2 Emission Grams/Mile', y='Highway MPG', hue='Drivetrain', data=data)
```

```python
#relplot: Facetgrid

sns.relplot(x='CO2 Emission Grams/Mile', y='Highway MPG', hue='Fuel Type', data=data, size= 'Fuel Type')
```

```python
sns.catplot(x='Highway MPG', y='Drivetrain', hue='Fuel Type', data=data)
```

## EDA Charts

```python
vehicles = pd.read_csv('/Users/ljgoico/desktop/ironhack/module-2/week-10/vehicles.csv')
```

```python
vehicles.head()
```

```python
vehicles.boxplot(column = 'Combined MPG')
```

```python
sns.boxplot(x='Fuel Type', y='Highway MPG', data=vehicles)
```

```python
#para arreglar que se vea todo ajuntao
ax = sns.boxplot(x='Fuel Type', y='Highway MPG', data=vehicles)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
cyl = [6.0,4.0]
vehicles.Cylinders.isin(cyl)
data1 = data[data.Cylinders.isin(cyl)]
data1.shape
```

```python
ax = sns.boxplot(x='Fuel Type', y='Highway MPG', data=data1, hue='Cylinders', palette="rocket")
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```
