### 2. Relación entre variables

### 2.1 Correlación

#### La correlación mide si una variable es linealmente dependiente de otra.
#### Ejemplo: y = ax+b

#### Si tengo correlación positiva, ambas variables crecen. Si es negativa, una crece y el otro decrece.

#### En la correlación, tenemos un rango de -1 a 1. Cuanto más me acerco al 0 por un lado u otro, menos correlación lineal hay.

#### La correlación es sensible a los outliers. 
#### Puede ser que no haya una relación lineal entre dos variables.
#### Las casualidades no se reflejarán en las correlaciones, puesto que la correlación debe ser potente.
#### La correlación estándar es la correlación de Fisher.
#### La correlación de Spearman y Kendall's Tau son otros tipos de correlaciones.

### 2.2 Covarianza

#### La covarianza es la medida de tendencia de dos variables a variar juntas.
#### Buscar info acerca de la fórmula y de cómo se calcula.
#### De la misma forma que tenemos una matriz de correlación, existe una matriz de covarianza, aunque ésta no tiene un rango específico. Cuanto más cercano a 0, menos covarianza hay.
