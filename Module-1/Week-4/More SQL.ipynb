```python
#HAVING siempre se pone después del group by
```

```python
QUERY = '''
SELECT 
    name,
    gender,
    SUM(number) AS total
FROM `bigquery-public-data.usa_names.usa_1910_2013`
GROUP BY
    name,
    gender
HAVING
    SUM(number)>40000
ORDER BY 
    total DESC
LIMIT
    10
'''
```

```python
query_job = client.query(QUERY)

df = query_job.to_dataframe()

df.head(10)
```

```python
#precio medio para cada tipo usando la tabla de títulos donde la fecha de publicación de 1991
QUERY = '''
SELECT 
    type,
    AVG(price) AS average_price
FROM `ironhack-data-analytics.publications.titles`
WHERE
    EXTRACT(YEAR FROM pubdate) > 1991
GROUP BY
    type
HAVING
    AVG(price)>0
ORDER BY 
    average_price DESC

'''
```

```python
query_job = client.query(QUERY)

df = query_job.to_dataframe()

df.head(10)
```

```python
#NÚMERO DE PUBLICACIONES POR PUBLISHER
```

```python
QUERY = '''
SELECT 
    publishers.pub_id,
    pub_name,
    COUNT(title) AS pub_per_pubs
FROM `ironhack-data-analytics.publications.publishers` publishers
INNER JOIN `ironhack-data-analytics.publications.titles` titles
ON publishers.pub_id = titles.pub_id
GROUP BY
    1,2
'''
```

```python
query_job = client.query(QUERY)

df = query_job.to_dataframe()

df.head(10)
```
