```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://www.reuters.com/article/us-shazam-m-a-apple-eu/eu-clears-apples-purchase-of-shazam-idUSKCN1LM1TZ'
```

```python
#para obtener el contenido. es decir: el .content te saca el html

html = requests.get(url).content

#como no lo quiero todo, se pueden hacer las primeras 600 palabras, por ejemplo

html[0:600]
```

```python
#Beautiful Soup hace algo que se llama parsing, que quita cosas para que sea más legible

soup = bs(html, 'lxml')

#esto te da todo el html

soup
```

```python
#ahora queremos encontrar todos los títulos y todos los párrafos.
#vamos a crear una lista llamada tags, porque son los tags que queremos de nuestro html

tags = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'p']
```

```python
#ahora le decimos: encuentra todos estos elementos en nuestro soup

text = [element.text for element in soup.find_all(tags)]

text
```

```python
#sale un poco feo. en la clase de string utilizamos algo que era join, y así se ve más claro.

print('\n'.join(text))
```

#### Ahora vamos a buscar una tabla en una página de wikipedia 

```python
url = 'https://en.wikipedia.org/wiki/List_of_European_countries_by_life_expectancy'

html = requests.get(url).content
soup = bs(html, 'lxml')
```

```python
#solo queremos obtener la información que se refiere a esa tabla. los parámetros son el tag y
#la clase de cómo se llama la tabla, en forma de diccionario

table = soup.find_all('table', {'class': 'sortable wikitable'})[0]

table
```

```python
#como el tr son rows, creamos lo siguiente:

rows = table.find_all('tr')
rows
```

```python
#ahora separamos basado en los parámetros que pongamos. Strip separa las palabras y split \n separa por saltos de línea

rows = [row.text.strip().split('\n') for row in rows]
rows
```

```python
#nos cargamos los strings vacíos con un list comprehension que lo que hace es filtrar

rows = [list(filter(None,row)) for row in rows]
rows
```

```python
#ahora transformamos esta lista en un dataframe

colnames = ['Rank', 'Country', 'Life expectancy']

df = pd.DataFrame(rows, columns = colnames)
df.head()
```

```python
#como se duplican el nombre de las columnas, lo seleccionamos así

df = pd.DataFrame(rows[1:], columns = colnames)
df.head()
```
